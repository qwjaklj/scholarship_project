### Quick Look at Data

```python
import pandas as pd

nearterm_data = pd.read_csv('../data/nearterm_data_2020-2024.csv')
historic_data = pd.read_csv('../data/NABR_historic.csv')

nearterm_data.head()
```

### Drop Wrong Columns

```python
historic_data.drop(columns = ['PPT_Annual','T_Annual'],inplace=True)
print(historic_data.shape)
historic_data.drop_duplicates(inplace=True)
print(historic_data.shape)

nearterm_data.drop(columns = ['PPT_Annual','T_Annual'],inplace=True)
print(nearterm_data.shape)
nearterm_data.drop_duplicates(inplace=True)
print(nearterm_data.shape)
```

```python
historic_data.columns
```

### Fill Empty Data Points

```python
# Make Plant Coverate be percentage rather than a number
historic_data['treecanopy_percent'] = 100 * historic_data['treecanopy']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])
historic_data['Ann_Herb_percent'] = 100 * historic_data['Ann_Herb']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])
historic_data['Bare_percent'] = 100 * historic_data['Bare']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])
historic_data['Herb_percent'] = 100 * historic_data['Herb']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])
historic_data['Litter_percent'] = 100 * historic_data['Litter']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])
historic_data['Shrub_percent'] = 100 * historic_data['treecanopy']/(historic_data['treecanopy']+ historic_data['Ann_Herb'] + historic_data['Bare']+ historic_data['Herb']+ historic_data['Litter'] + historic_data['Shrub'])

nearterm_data['treecanopy_percent'] = 100 * nearterm_data['treecanopy']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])
nearterm_data['Ann_Herb_percent'] = 100 * nearterm_data['Ann_Herb']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])
nearterm_data['Bare_percent'] = 100 * nearterm_data['Bare']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])
nearterm_data['Herb_percent'] = 100 * nearterm_data['Herb']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])
nearterm_data['Litter_percent'] = 100 * nearterm_data['Litter']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])
nearterm_data['Shrub_percent'] = 100 * nearterm_data['treecanopy']/(nearterm_data['treecanopy']+ nearterm_data['Ann_Herb'] + nearterm_data['Bare']+ nearterm_data['Herb']+ nearterm_data['Litter'] + nearterm_data['Shrub'])

# Drop unnecessary columns
historic_data.drop(columns=['treecanopy','Ann_Herb', 'Bare', 'Herb', 'Litter', 'Shrub'],inplace=True)
nearterm_data.drop(columns=['treecanopy','Ann_Herb', 'Bare', 'Herb', 'Litter', 'Shrub'],inplace=True)

# Round Numbers
historic_data = historic_data.round(3)
nearterm_data = nearterm_data.round(3)

# Code refers to https://saturncloud.io/blog/how-to-combine-multiple-rows-into-a-single-row-with-pandas/#:~:text=One%20way%20to%20combine%20multiple,aggregation%20function%20to%20each%20group.
agg_functions = {'DrySoilDays_Summer_whole': 'sum', 'Evap_Summer' : 'sum',
    'ExtremeShortTermDryStress_Summer_whole' : 'sum', 'FrostDays_Winter' : 'sum',
    'NonDrySWA_Summer_whole' : 'sum', 'PPT_Winter' : 'sum', 'PPT_Summer' : 'sum', 'T_Winter': 'sum',
    'T_Summer': 'sum', 'Tmax_Summer': 'sum', 'Tmin_Winter': 'sum', 'VWC_Winter_whole': 'sum',
    'VWC_Spring_whole': 'sum', 'VWC_Summer_whole': 'sum', 'VWC_Fall_whole': 'sum' }

required_columns = ['long', 'lat', 'year', 'TimePeriod', 'RCP', 'scenario', 'treecanopy_percent',
    'Ann_Herb_percent', 'Bare_percent', 'Herb_percent', 'Litter_percent', 'Shrub_percent']

historic_data = historic_data.groupby(required_columns).aggregate(agg_functions)
nearterm_data = nearterm_data.groupby(required_columns).aggregate(agg_functions)

historic_data.to_csv('../data/historic_cleaned.csv')
nearterm_data.to_csv('../data/nearterm_cleaned.csv')
```

### Data Merging

```python
nearterm_df = pd.read_csv('../data/nearterm_cleaned.csv')
historic_df = pd.read_csv('../data/historic_cleaned.csv')
merged_data = pd.concat([nearterm_df, historic_df], ignore_index=True)
#print(merged_data.columns)
merged_data = merged_data.sort_values(by='year').groupby(['long', 'lat']).apply(lambda x: x).reset_index(drop=True)
merged_data = merged_data.round(3)
merged_data.head()
```
